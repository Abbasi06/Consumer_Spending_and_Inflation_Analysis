# **Investigating United States Consumer's Income, Expenditure, and Inflation**

```python
#Importing all the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
```

## **United States Personal Income & Outlays**

```python
# Reading the US personal income & outlays file for the 2021,2022 & 2023 in quarterwise
income_df = pd.read_excel("/content/Personal Income_Quarterly.xlsx", sheet_name = "Income")
income_df
```

```python
pd.set_option("display.max_rows" , 10) #setting the pandas option to display only 10 rows
```

```python
# Changing the column names
income_df.columns= ["main_category", "category", "sub_category", "items", "2021_Q1", "2021_Q2", "2021_Q3", "2021_Q4", "2022_Q1", "2022_Q2", "2022_Q3", "2022_Q4", "2023_Q1", "2023_Q2", "2023_Q3", "2023_Q4"]
income_df
```

```python
income_df.info() #dataset doesnt have any null values
```

```python
income_df.describe() # Describing the income dataframe
```

```python
# Applying groupby function over below categories
Total_df = income_df[(income_df != "Population") ].groupby("main_category")[['2021_Q1','2021_Q2', '2021_Q3', '2021_Q4', '2022_Q1', '2022_Q2', '2022_Q3','2022_Q4', '2023_Q1', '2023_Q2', '2023_Q3', '2023_Q4']].sum().sort_values(by = "2021_Q1", ascending = False).reset_index()
Total_df
```

```python
# Transporting the dataframe to get the desired dataframe
transposed_total_df = Total_df.T.reset_index()
transposed_total_df.columns = transposed_total_df.iloc[0]
transposed_total_df.drop(0, inplace = True)
display(transposed_total_df)
```

```python
# Disposable Personal Income (Personal Income - Deductions)
transposed_total_df["disposable_income"] = transposed_total_df["Personal Income"].astype(float) - transposed_total_df["Deductions"].astype(float)
transposed_total_df["year"] = [2021, 2021, 2021, 2021, 2022, 2022, 2022, 2022, 2023, 2023, 2023, 2023]
transposed_total_df["quarters"] = ["Q1", "Q2", "Q3", "Q4", "Q1", "Q2", "Q3", "Q4","Q1", "Q2", "Q3", "Q4"]
transposed_total_df
```

```python
# How much an average disposable income available from the personal income yearly? & finding the disposable income percentage and savings percentage
average_dpi = transposed_total_df.groupby("year")["disposable_income"].mean().reset_index().set_index("year")
average_dpi["total_income"] = transposed_total_df.groupby("year")["Personal Income"].mean()
average_dpi["personal_expenses"] = transposed_total_df.groupby("year")["Expenditures"].mean()
average_dpi["tax_deductions"] = transposed_total_df.groupby("year")["Deductions"].mean()
average_dpi["savings"] =  transposed_total_df.groupby("year")["Savings"].mean()
average_dpi["disposable_income_percentage"] = ((transposed_total_df.groupby("year")["disposable_income"].mean()) / (transposed_total_df.groupby("year")["Personal Income"].mean())) * 100
average_dpi = average_dpi[['total_income', "tax_deductions", 'disposable_income', 'disposable_income_percentage', 'personal_expenses', 'savings']]
average_dpi["savings_percentage"] =  (transposed_total_df.groupby("year")["Savings"].mean()/ average_dpi.disposable_income) * 100
display(average_dpi)

#Income and Expenses Over Time
years = average_dpi.index
tax_deductions = average_dpi.tax_deductions
personal_expenses = average_dpi.personal_expenses
savings = average_dpi.savings
trace2 = go.Bar(x=years, y=tax_deductions, name='Tax Deductions')
trace3 = go.Bar(x=years, y=personal_expenses, name='Personal Expenses')
trace4 = go.Bar(x=years, y=savings, name='Savings')

layout = go.Layout(
    title='Income and Expenses Over Time',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Value in Billions'),
    barmode='stack'
)
fig = go.Figure(data=[trace2, trace3, trace4], layout=layout)
fig.show()
```

```python
# Total Personal Income Categories

income_category_df = income_df[income_df.main_category == "Personal Income"].groupby("category")[['2021_Q1','2021_Q2', '2021_Q3', '2021_Q4', '2022_Q1', '2022_Q2', '2022_Q3','2022_Q4', '2023_Q1', '2023_Q2', '2023_Q3', '2023_Q4']].sum()
income_category_df
```

```python
# Employee Compensation Percentage
employee_compensation_percentage = (income_category_df.iloc[0]/income_category_df.sum()) * 100
print("Average Employee Compensation Percentage: " ,employee_compensation_percentage.mean())
```

# **United States Personal Consumption Expenditure**

```python
#Reading Expenditures File
expense_df = pd.read_excel("/content/Personal Consumption Expenditures_Aggregate in Dollars.xlsx", sheet_name = "Expenses")
expense_df
```

```python
# Applying groupby function over expenditure categories
expense_df1 = expense_df.groupby(["Main-Category", "Category"])[['2021 Q1','2021 Q2', '2021 Q3', '2021 Q4', '2022 Q1', '2022 Q2', '2022 Q3','2022 Q4', '2023 Q1', '2023 Q2', '2023 Q3', '2023 Q4']].sum().reset_index()
expense_df1.columns = ['main-Category', 'category', '2021_Q1', '2021_Q2', '2021_Q3', '2021_Q4', '2022_Q1', '2022_Q2', '2022_Q3', '2022_Q4', '2023_Q1', '2023_Q2', '2023_Q3', '2023_Q4']
display(expense_df1)
```

```python
#Converting Object type to float
for i in expense_df1.columns[2:]:
  expense_df1[i].astype(float)
expense_df1.info()
```

```python
# Personal Consumption Expenditure Quarterwise difference
columns = expense_df1.columns[2:]
expense_df1_diff = pd.DataFrame()
try:
  for i in range(len(columns)):
    expense_df1_diff[f"{columns[i+1]} - {columns[i]}"] = ((expense_df1[f"{columns[i+1]}"] - expense_df1[f"{columns[i]}"])/ expense_df1[f"{columns[i]}"]) * 100
except:
    expense_df1_diff
print()
print("Personal Consumption Expenditure Quarterwise difference: ")
display(expense_df1_diff)
print("Goods are rising at an mean percentage of :" ,expense_df1_diff.iloc[0].mean())
print("Services are rising at an mean percentage of :", expense_df1_diff.iloc[1].mean())
```

```python
expense_df1_T = expense_df1.T.reset_index()
expense_df1_T.columns = ["Quarter","Goods", "Services"]
expense_df1_T.drop(0, inplace = True)
expense_df1_T.drop(1, inplace = True)
expense_df2 = expense_df1_T.reset_index()
expense_df2.drop(columns = ["index"], inplace = True)
expense_df2

# Trend of goods over quarters
expense_df_items = expense_df[expense_df["Category"] == "Goods"][['Items', '2021 Q1',
       '2021 Q2', '2021 Q3', '2021 Q4', '2022 Q1', '2022 Q2', '2022 Q3',
       '2022 Q4', '2023 Q1', '2023 Q2', '2023 Q3', '2023 Q4']]
print("List of all goods over quarters :")
display(expense_df_items)
print()
plt.figure(figsize=(12, 6))


for item in expense_df_items['Items']:
    plt.plot(expense_df_items.columns[1:], expense_df_items[expense_df_items['Items'] == item].iloc[0, 1:], label=item, linewidth = 1.5)

plt.xlabel('Quarter')
plt.ylabel('Amount in Billions')
plt.title('Trends of Goods over Quarters')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()

# Trend of services over quarters
expense_df_items = expense_df[expense_df["Category"] == "Services"][['Items', '2021 Q1',
       '2021 Q2', '2021 Q3', '2021 Q4', '2022 Q1', '2022 Q2', '2022 Q3',
       '2022 Q4', '2023 Q1', '2023 Q2', '2023 Q3', '2023 Q4']]
print("List of all services over quarters :")
display(expense_df_items)
print()
plt.figure(figsize=(12, 6))


for item in expense_df_items['Items']:
    plt.plot(expense_df_items.columns[1:], expense_df_items[expense_df_items['Items'] == item].iloc[0, 1:], label=item, linewidth = 1.5)

plt.xlabel('Quarter')
plt.ylabel('Amount in Billions')
plt.title('Trends of services over Quarters')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

# **United States Consumer Price Index(CPI)**

```python
# Reading all the CPI files

gas = pd.read_csv('/content/Gasoline_cpi.csv', usecols=['Year', 'Label', 'Value'])
bread = pd.read_csv('/content/Bread_cpi.csv', usecols=['Year', 'Label', 'Value'])
milk = pd.read_csv('/content/Milk_cpi.csv', usecols=['Year', 'Label', 'Value'])
current = pd.read_csv('/content/Electricity_cpi.csv', usecols=['Year', 'Label', 'Value'])

new_columns = {'Year': 'year', 'Label': 'month', 'Value': 'value'}

gas.rename(columns=new_columns, inplace=True)
bread.rename(columns=new_columns, inplace=True)
milk.rename(columns=new_columns, inplace=True)
current.rename(columns=new_columns, inplace=True)

# Aggregate data to represent it year-wise
gas_yearly_avg = gas.groupby('year')['value'].mean().reset_index()
bread_yearly_avg = bread.groupby('year')['value'].mean().reset_index()
milk_yearly_avg = milk.groupby('year')['value'].mean().reset_index()
current_yearly_avg = current.groupby('year')['value'].mean().reset_index()


fig = go.Figure()
fig.add_trace(go.Scatter(x=gas_yearly_avg['year'], y=gas_yearly_avg['value'],
                         mode='lines+markers', name='Gas'))
fig.add_trace(go.Scatter(x=bread_yearly_avg['year'], y=bread_yearly_avg['value'],
                         mode='lines+markers', name='Bread'))
fig.add_trace(go.Scatter(x=milk_yearly_avg['year'], y=milk_yearly_avg['value'],
                         mode='lines+markers', name='Milk'))
fig.add_trace(go.Scatter(x=current_yearly_avg['year'], y=current_yearly_avg['value'],
                         mode='lines+markers', name='Electricity'))
fig.update_layout(title='Comparison of Yearly Datasets (Average)',
                  xaxis_title='Year', yaxis_title='Average CPI Value')
fig.update_xaxes(tickmode='linear', dtick=1)

fig.show()
```

```python
# Reading other CPI file

df = pd.read_csv('/content/Consumers_CPI.csv')
df_long = pd.melt(df, id_vars=['year'], var_name='month', value_name='value')
yearly_values = df_long.groupby('year')['value'].mean().reset_index()
print("Yearly-wise CPI Index: ")
display(df)

fig = px.line(df_long, x='month', y='value', color='year',
              title='CPI Yearly Trends Over Months, [Index numbers, 2017=100], Seasonally adjusted',
              labels={'month': 'Month', 'value': 'Value', 'year': 'Year'})
fig.update_layout(xaxis_title='Month', yaxis_title='Value')
fig.show()
```

```python
cpi_df = df.copy()
cpi_df.drop(cpi_df.iloc[-1].name, inplace = True)
cpi_df = cpi_df.tail(3)
display(cpi_df) #Considering only the last three year data

cpi_df["Q1"] = (cpi_df["Jan"] + cpi_df["Feb"] + cpi_df["Mar"])/3
cpi_df["Q2"] = (cpi_df["Apr"] + cpi_df["May"] + cpi_df["June"])/3
cpi_df["Q3"] = (cpi_df["July"] + cpi_df["Aug"] + cpi_df["Sep"])/3
cpi_df["Q4"] = (cpi_df["Oct"] + cpi_df["Nov"] + cpi_df["Dec"])/3
cpi_df1 = cpi_df[["year", "Q1", "Q2", "Q3", "Q4"]]
cpi_df2 = cpi_df1.T.reset_index()
cpi_df2.columns = ["Quarter", "2021", "2022", "2023"]
cpi_df2.drop(0 , inplace = True)
display(cpi_df2) #Segregating monthly CPI data to Quarterly CPI data
```

```python
# Melt the dataframe to reshape it
melted_df = cpi_df2.melt(id_vars=['Quarter'], var_name='Year', value_name='Value')

# Sort the values by 'Year' and 'Quarter'
# melted_df['Year'] = melted_df['Year'].astype(int) # Convert 'Year' column to integer for correct sorting
melted_df.sort_values(by=['Year', 'Quarter'], inplace=True)

# Reset the index
melted_df.reset_index(drop=True, inplace=True)
melted_df["Qtr_yr"] = melted_df["Year"] + "_" +melted_df["Quarter"]

melted_df.columns = ["Quarter", "Year", "Average_CPI", "Qtr_yr"]
melted_df
```

```python
final_df = pd.merge(merged_df2, melted_df, left_on = "year_qtr", right_on = "Qtr_yr", how = "left")
final_df = final_df[["year_qtr", "year", "quarter", "personal_income", "disposable_income", "savings", "goods", "services", "Average_CPI"]]
final_df # performing merge operation to get income, expenses and Avergae CPI data
```

```python
# Reading PCE index files
pce_index_df = pd.read_excel("/content/Personal Consumption Expenditures_Index.xlsx", sheet_name = "PCE_Index")
pce_index_df
```

```python
pce_index_df.columns = ["main_category", "category", "sub-category", "items", "2021_Q1",'2021_Q2', '2021_Q3', '2021_Q4', '2022_Q1', '2022_Q2', '2022_Q3',
       '2022_Q4', '2023_Q1', '2023_Q2', '2023_Q3', '2023_Q4']
pce_index_df1 = pce_index_df.loc[1:].groupby("category")[["2021_Q1",'2021_Q2', '2021_Q3', '2021_Q4', '2022_Q1', '2022_Q2', '2022_Q3',
       '2022_Q4', '2023_Q1', '2023_Q2', '2023_Q3', '2023_Q4']].mean()
display(pce_index_df1) #applying groupby operation to get average pce index for goods and services
```

```python
pce_index_df2 = pce_index_df1.T.reset_index()
pce_index_df2.columns = ["Quarter", "Goods_PCE_Index", "Services_PCE_Index"]
pce_index_df2 # Transposing the dataframe
```

```python
final_df1 = pd.merge(final_df, pce_index_df2, left_on = "year_qtr", right_on = "Quarter" , how = "left")
final_df1 # merging all the income, expenditure, Average CPI and PCE index
```
